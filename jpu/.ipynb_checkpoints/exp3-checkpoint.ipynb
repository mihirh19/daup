```python
df['airconditioning']=df['airconditioning'].replace({'no':'0','yes':'1'})
df['airconditioning']
```

```python
df['furnishingstatus']=df['furnishingstatus'].str.replace('semi-furnished','0.5')
df['furnishingstatus']=df['furnishingstatus'].str.replace('unfurnished','0')
df['furnishingstatus']=df['furnishingstatus'].str.replace('furnished','1')
df
```

```python
df['mainroad'].str.contains('yes')
df['mainroad'].str.contains('no')
```

```python
df['mainroad']=df['mainroad'].str.replace('no','0')
df['mainroad']=df['mainroad'].str.replace('yes','1')
df
```

```python
df=df.fillna( method='bfill')
```

```python
guestroom_group = df.groupby('guestroom')
guestroom_mean = guestroom_group['parking'].mean()
guestroom_mean
```

```python
df['parkingstatus']=np.where(df['parking']==0,'nopark',df['parking'])

df
```

```python
df['guestroom-basement']=df['guestroom']+' '+df['basement']
```

```python
df
```

```python
pivot_df=df.pivot_table(index='area',values='price',aggfunc=np.mean)
```
