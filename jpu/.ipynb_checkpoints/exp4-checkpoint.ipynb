### Load Housing Data set

```python
df  = pd.read_csv('Housing.csv')
```

```python
df
```

```python
plt.plot(df['price'])
```

```python
plt.scatter(x=df['price'], y=df['area'], s=5)
plt.xlabel('price')
plt.ylabel('area')
plt.title('price vs area')
```

```python
plt.boxplot([df['price'],df['area']])
plt.title('box Chart of Categories')
plt.show()
```

```python
df_unique=df['bedrooms'].unique()
df_unique
```

```python
import numpy as np
categories = np.array(df_unique)
bedroom_counts = df['bedrooms'].value_counts()

# Create a pie chart
plt.pie(bedroom_counts, labels=bedroom_counts.index, autopct=lambda pct: f"{pct:.1f}%\n({int(pct/100*sum(bedroom_counts))})")
plt.title('Distribution of Bedrooms')
plt.show()
```
