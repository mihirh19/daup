<h4> Aim : To perform KMeans and DBSCAN clustering using sklearn package in python.</h4>

Generate Random Data for clustering

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, y_true = make_blobs(n_samples=100, centers=3, cluster_std=0.6)
plt.scatter(X[:,0],X[:,1], s=50)
```

Create KMeans Clustering Model

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
```

Run Clustering Algorithm to fit the data and find cluster labels

```python
km.fit(X)
y_lab = km.predict(X)
```

Plot the data points using scatter-plot

```python
plt.scatter(X[:, 0], X[:, 1], c=y_lab, s=50, cmap='viridis')
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200,
alpha=0.5);
```

Create DBSCAN model

```python
from sklearn.cluster import dbscan
dbs = dbscan(X)
```

Plot Data Points

```python
plt.scatter(X[:, 0], X[:, 1], c=dbs[1], s=50, cmap='viridis')
```
