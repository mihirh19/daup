```python
from textblob import TextBlob
```

Create a TextBlob Object from Text

```python
text_data = "This is a simple text data for textblob to process and analyze. It is a very simple text data."

blob = TextBlob(text_data)

blob.correct() #will correct the spelling mistakes if any.
```

Part of Speech Tagging (POS Tags)

```python
print("Tokens : ", blob.words)
print("Sentiment : ", blob.sentiment)
print("tags : ", blob.tags)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud().generate(text_data)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
